```python
import matplotlib.pyplot as plt
from pathlib import Path
import csv
import seaborn as sns
sns.set()

# import required packages 
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
# addressing the path of file
path = Path('datasets/global_tourist_hotspot/cleaned_data_Iran.csv')
#"C:\Users\Dariush\Desktop\python_work\python-projects\data-visualization-project\datasets\global_tourist_hotspot\cleaned_data_Iran.csv
```

```python
# reading csv file
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)
print(header_row)
```

```python
# printing headers
for index, column_header in enumerate(header_row):
    print(index, column_header)
```

```python
rates = []
names = []
reviews = []
weighted_scores = []
for row in reader:
    name = row[1]
    names.append(name)
    rate = float(row[3])
    rates.append(rate)
    review = row[4]
    reviews.append(review)
    weighted_score = float(row[12])
    weighted_scores.append(weighted_score)
    
print(rates)
```

```python
fig, ax = plt.subplots()

ax.scatter(names, rates, color='red')
ax.set_title("Tourism Rate in Iran", fontsize=24)
ax.set_xlabel("place name", fontsize=8)
ax.set_ylabel("Tourism Rate", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
fig = px.bar(x = names, y = weighted_scores, title = "Tourism score in Iran", labels = {'x': 'place name', 'y': 'Tourism Rate'})
fig.show()
```
