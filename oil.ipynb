```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
sns.set()
import csv

# import required packages 
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'

from datetime import datetime
```

```python
# addressing the path of file
path = Path('datasets/opec/oil_production.csv')
```

```python
# reading csv file
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)
print(header_row)
```

```python
# printing headers
for index, column_header in enumerate(header_row):
    print(index, column_header)
```

```python
# extracting volume of production
dates, barrels = [], []
for row in reader:
    try:
        barrel = float(row[1])
        current_date = datetime.strptime(row[0], '%Y-%m')
        
    except:
        print(f'missing data for {row[0]}')
    else:
        barrels.append(barrel)
        dates.append(current_date)
    
print(barrels)
print(dates)
```

```python
fig, ax = plt.subplots(figsize = (16,9))
ax.plot(dates, barrels, color = 'brown', linewidth = 2)

ax.set_title('monthly Oil production of OPEC')
ax.set_xlabel('index')
ax.set_ylabel('number of barrels in million')

fig.autofmt_xdate()

plt.show()
```
